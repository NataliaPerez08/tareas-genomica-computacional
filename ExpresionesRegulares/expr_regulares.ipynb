# Expresiones regulares (ejercicio)


Encontrar y contar todas las secuencias que pertenezcan a una misma etiqueta usando expresiones regulares
ejemplo: Las secuencias etiquetadas como

PZ7180000000004_TX
PZ7180000000004_TY
PZ7180000000004_TZ

pertenecen todas a PZ7180000000004

PZ7180000000067_AF
PZ7180000000067_AY

pertenecen ambas a PZ7180000000067

El script que generen debe obtener el identificador asociado y cuantas secuencias caen bajo ese identifciador
Para nuestras dos identificadores:

PZ7180000000067    2
PZ7180000000004    3

```python
path = "./pz_cDNAs.fasta" 

secuencia = ""
with open(path, "r") as file:
    for line in file:
        secuencia += line

secuencia.split("\n")[0:5]
```

```python
diccionario = dict()
for i in secuencia.split("\n"):
    if i.startswith(">"):
        etiqueta = i[1:].split(" ")[0].split("_")[0]
        if etiqueta in diccionario:
            diccionario[etiqueta] += 1
        else:
            diccionario[etiqueta] = 1
```

```python
print("PZ7180000000004  aparece", diccionario["PZ7180000000067"], "veces")
print("PZ7180000000067  aparece", diccionario["PZ7180000000004"], "veces")
```

```python
# Diccioanrio con las etiquetas y el n√∫mero de veces que aparece
print("Diccionario con las etiquetas y el n√∫mero de veces que aparece")
print(diccionario)
```
